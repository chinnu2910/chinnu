```python
import matplotlib.pyplot as plt
import seaborn as sns

# Numerical features analysis
numerical_features = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary', 'Tenure', 'NumOfProducts']
for col in numerical_features:
    churn_stats = df.groupby('Exited')[col].agg(['mean', 'median'])
    print(f"Statistics for {col} grouped by Exited:\n{churn_stats}")

# Categorical features analysis
categorical_features = ['Geography', 'Gender', 'HasCrCard', 'IsActiveMember']
plt.figure(figsize=(12, 8))
for i, col in enumerate(categorical_features):
    plt.subplot(2, 2, i + 1)
    churn_rates = df.groupby(col)['Exited'].mean()
    sns.barplot(x=churn_rates.index, y=churn_rates.values)
    plt.title(f'Churn Rate by {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('churn_rate_by_categorical_features.png')
plt.show()


# Interaction effects (example: Age and Exited across Geography)
interaction_table = df.groupby(['Geography', 'Exited'])['Age'].mean().unstack()
print("\nInteraction effect of Age and Exited across Geography:\n", interaction_table)
```
